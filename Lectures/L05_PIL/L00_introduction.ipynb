## 주요 이미지 처리 라이브러리

### 1. **PIL (Python Imaging Library)**
- 1995년 개발된 최초의 Python 이미지 처리 라이브러리
- 2009년 이후 업데이트 중단 → 현대 Python과 호환성 문제 발생

### 2. **Pillow**
- PIL의 현대화된 포크 버전 (2010년~)
- PIL과 동일한 API 제공하며 지속적으로 업데이트
- 모듈명은 여전히 `PIL` (설치는 `pillow`)
- **특징**: OOP 방식, 사용 간편, Django/PyTorch 등 다른 패키지와 호환성 우수

```python
from PIL import Image
pil_image = Image.open("image.png")
```

### 3. **OpenCV (cv2)**
- 2000년 Intel 개발, 컴퓨터 비전 특화
- 이미지 처리 + 비디오 분석, 얼굴 인식, 객체 추적 등
- **특징**: 실시간 처리 강력, NumPy 배열 기반, 가장 풍부한 기능

### 4. **scikit-image**
- 2009년 시작, SciKits 생태계의 일부
- NumPy 기반, 과학 연구 및 데이터 분석에 최적화
- **특징**: 연구 논문 구현체가 많음

## 선택 가이드
- **간단한 이미지 작업, 다른 패키지와 통합**: Pillow
- **컴퓨터 비전, 실시간 처리**: OpenCV
- **과학적 분석, 연구**: scikit-image
