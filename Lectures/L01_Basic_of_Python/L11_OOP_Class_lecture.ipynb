### 1.3 Methodì˜ ì¢…ë¥˜

#### 1. Instance Method (ì¸ìŠ¤í„´ìŠ¤ ë©”ì„œë“œ)
- **ì •ì˜**: `self`ë¥¼ ì²« ë²ˆì§¸ ë§¤ê°œë³€ìˆ˜ë¡œ ë°›ëŠ” ë©”ì„œë“œ
- **ìš©ë„**: Instanceì˜ ì†ì„±ì— ì ‘ê·¼í•˜ê±°ë‚˜ ìˆ˜ì •
- **í˜¸ì¶œ**: `instance.method()`

#### 2. Class Method (í´ë˜ìŠ¤ ë©”ì„œë“œ)
- **ì •ì˜**: `@classmethod` ë°ì½”ë ˆì´í„° ì‚¬ìš©, `cls`ë¥¼ ì²« ë²ˆì§¸ ë§¤ê°œë³€ìˆ˜ë¡œ ë°›ìŒ
- **ìš©ë„**: Class Attributeì— ì ‘ê·¼, Alternative Constructor (ëŒ€ì²´ ìƒì„±ì)
- **í˜¸ì¶œ**: `Class.method()` ë˜ëŠ” `instance.method()`

#### 3. Static Method (ì •ì  ë©”ì„œë“œ)
- **ì •ì˜**: `@staticmethod` ë°ì½”ë ˆì´í„° ì‚¬ìš©, `self`ë‚˜ `cls` ì—†ìŒ
- **ìš©ë„**: Classì™€ ê´€ë ¨ì€ ìˆì§€ë§Œ Instanceë‚˜ Class Attributeì— ì ‘ê·¼ ë¶ˆí•„ìš”í•œ ê²½ìš°
- **í˜¸ì¶œ**: `Class.method()` ë˜ëŠ” `instance.method()`

````python
# Methodì˜ ì¢…ë¥˜
print("=== Method ì¢…ë¥˜ ===")

class Person:
    population = 0  # Class Attribute
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person.population += 1
    
    # 1. Instance Method
    def introduce(self):
        return f"ì•ˆë…•í•˜ì„¸ìš”, ì €ëŠ” {self.name}ì´ê³  {self.age}ì‚´ì…ë‹ˆë‹¤."
    
    def birthday(self):
        self.age += 1
        print(f"{self.name}ì˜ ìƒì¼! ì´ì œ {self.age}ì‚´")
    
    # 2. Class Method (ëŒ€ì²´ ìƒì„±ì)
    @classmethod
    def from_birth_year(cls, name, birth_year):
        age = 2025 - birth_year
        return cls(name, age)
    
    @classmethod
    def get_population(cls):
        return f"í˜„ì¬ ì¸êµ¬: {cls.population}ëª…"
    
    # 3. Static Method
    @staticmethod
    def is_adult(age):
        return age >= 18
    
    @staticmethod
    def calculate_age(birth_year):
        return 2025 - birth_year

# Instance Method ì‚¬ìš©
p1 = Person("ì² ìˆ˜", 25)
print(p1.introduce())
p1.birthday()

# Class Method ì‚¬ìš© (ëŒ€ì²´ ìƒì„±ì)
p2 = Person.from_birth_year("ì˜í¬", 2000)
print(p2.introduce())

# Class Method ì‚¬ìš© (Class Attribute ì ‘ê·¼)
print(Person.get_population())

# Static Method ì‚¬ìš©
print(f"\n20ì‚´ì€ ì„±ì¸? {Person.is_adult(20)}")
print(f"2000ë…„ìƒì˜ ë‚˜ì´: {Person.calculate_age(2000)}ì‚´")
````

````python
# Method ì¢…ë¥˜ ë¹„êµ ì •ë¦¬
print("\n=== Method ë¹„êµ ===")

class Example:
    class_var = "í´ë˜ìŠ¤ ë³€ìˆ˜"
    
    def __init__(self, value):
        self.instance_var = value
    
    def instance_method(self):
        # selfë¥¼ í†µí•´ Instanceì™€ Classì— ëª¨ë‘ ì ‘ê·¼ ê°€ëŠ¥
        return f"Instance: {self.instance_var}, Class: {self.class_var}"
    
    @classmethod
    def class_method(cls):
        # clsë¥¼ í†µí•´ Classì—ë§Œ ì ‘ê·¼ ê°€ëŠ¥
        return f"Class: {cls.class_var}"
        # return f"Instance: {cls.instance_var}"  # ì—ëŸ¬!
    
    @staticmethod
    def static_method(param):
        # selfë‚˜ cls ì—†ì´ ë…ë¦½ì ìœ¼ë¡œ ë™ì‘
        return f"Static: {param}"
        # return f"Instance: {self.instance_var}"  # ì—ëŸ¬!

ex = Example("ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜")

print("Instance Method:", ex.instance_method())
print("Class Method:", Example.class_method())
print("Static Method:", Example.static_method("íŒŒë¼ë¯¸í„°"))
````

### 2.2 ê¸°ë³¸ Special Methods

#### `__str__` vs `__repr__`

| Method | ìš©ë„ | í˜¸ì¶œ ë°©ë²• | ëª©ì  |
|--------|------|-----------|------|
| `__str__` | ì‚¬ìš©ìê°€ ë³´ê¸° ì¢‹ì€ ë¬¸ìì—´ | `str(obj)`, `print(obj)` | ê°€ë…ì„± (readable) |
| `__repr__` | ê°œë°œììš©, ê°ì²´ ì¬ìƒì„± ê°€ëŠ¥í•œ í‘œí˜„ | `repr(obj)`, ì½˜ì†”ì—ì„œ ì¶œë ¥ | ëª…í™•ì„± (unambiguous) |

**ì›ì¹™**:
- `__repr__`ëŠ” í•­ìƒ êµ¬í˜„í•˜ëŠ” ê²ƒì´ ì¢‹ìŒ
- `eval(repr(obj)) == obj`ê°€ ì´ìƒì  (ê°€ëŠ¥í•˜ë‹¤ë©´)
- `__str__`ì´ ì—†ìœ¼ë©´ `__repr__`ê°€ ëŒ€ì‹  ì‚¬ìš©ë¨

````python
# __str__ê³¼ __repr__ ë¹„êµ
print("=== __str__ vs __repr__ ===")

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def __str__(self):
        # ì‚¬ìš©ì ì¹œí™”ì  í‘œí˜„
        return f"{self.name} ({self.age}ì‚´)"
    
    def __repr__(self):
        # ê°œë°œììš©, ê°ì²´ ì¬ìƒì„± ê°€ëŠ¥í•œ í‘œí˜„
        return f"Person('{self.name}', {self.age})"

p = Person("ê¹€ì² ìˆ˜", 25)

print("str(p):", str(p))      # __str__ í˜¸ì¶œ
print("repr(p):", repr(p))    # __repr__ í˜¸ì¶œ
print("print(p):", end=" ")  
print(p)                       # __str__ í˜¸ì¶œ

# ì½˜ì†”ì—ì„œ ë³€ìˆ˜ë§Œ ì…ë ¥í•˜ë©´ __repr__ í˜¸ì¶œ
print("\nì½˜ì†”ì—ì„œ p ì…ë ¥ ì‹œ:")
print(repr(p))
````

````python
# __str__ë§Œ ìˆì„ ë•Œ vs ë‘˜ ë‹¤ ìˆì„ ë•Œ
print("\n=== __str__ê³¼ __repr__ ìœ ë¬´ ë¹„êµ ===")

class OnlyStr:
    def __init__(self, value):
        self.value = value
    
    def __str__(self):
        return f"Value: {self.value}"

class OnlyRepr:
    def __init__(self, value):
        self.value = value
    
    def __repr__(self):
        return f"OnlyRepr({self.value})"

obj1 = OnlyStr(10)
obj2 = OnlyRepr(20)

print("OnlyStr:")
print(f"  print(): {obj1}")
print(f"  repr(): {repr(obj1)}")

print("\nOnlyRepr:")
print(f"  print(): {obj2}")
print(f"  repr(): {repr(obj2)}")
````

### 2.3 ë¹„êµ ì—°ì‚°ì Special Methods

#### Comparison Operators

| ì—°ì‚°ì | Special Method | ì„¤ëª… |
|--------|---------------|------|
| `==` | `__eq__(self, other)` | Equal |
| `!=` | `__ne__(self, other)` | Not equal |
| `<` | `__lt__(self, other)` | Less than |
| `>` | `__gt__(self, other)` | Greater than |
| `<=` | `__le__(self, other)` | Less than or equal |
| `>=` | `__ge__(self, other)` | Greater than or equal |

**Tip**: `__eq__`ì™€ `__lt__`ë§Œ êµ¬í˜„í•˜ë©´ ë‚˜ë¨¸ì§€ëŠ” ìë™ìœ¼ë¡œ ì²˜ë¦¬ ê°€ëŠ¥ (`functools.total_ordering` ë°ì½”ë ˆì´í„° ì‚¬ìš©)

````python
# ë¹„êµ ì—°ì‚°ì êµ¬í˜„
print("=== ë¹„êµ ì—°ì‚°ì Special Methods ===")

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __eq__(self, other):
        """== ì—°ì‚°ì"""
        if not isinstance(other, Point):
            return False
        return self.x == other.x and self.y == other.y
    
    def __lt__(self, other):
        """< ì—°ì‚°ì (ì›ì ìœ¼ë¡œë¶€í„°ì˜ ê±°ë¦¬ ë¹„êµ)"""
        return (self.x**2 + self.y**2) < (other.x**2 + other.y**2)
    
    def __gt__(self, other):
        """> ì—°ì‚°ì"""
        return (self.x**2 + self.y**2) > (other.x**2 + other.y**2)
    
    def __repr__(self):
        return f"Point({self.x}, {self.y})"

p1 = Point(3, 4)  # ê±°ë¦¬: 5
p2 = Point(3, 4)  # ê±°ë¦¬: 5
p3 = Point(1, 2)  # ê±°ë¦¬: 2.24

print(f"p1 == p2: {p1 == p2}")
print(f"p1 == p3: {p1 == p3}")
print(f"p1 > p3: {p1 > p3}")
print(f"p3 < p1: {p3 < p1}")
````

````python
# functools.total_orderingì„ ì‚¬ìš©í•œ ìë™ ë¹„êµ
from functools import total_ordering

print("\n=== total_ordering ë°ì½”ë ˆì´í„° ===")

@total_ordering
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
    
    def __eq__(self, other):
        return self.grade == other.grade
    
    def __lt__(self, other):
        return self.grade < other.grade
    
    def __repr__(self):
        return f"Student('{self.name}', {self.grade})"

s1 = Student("ì² ìˆ˜", 85)
s2 = Student("ì˜í¬", 90)
s3 = Student("ë¯¼ìˆ˜", 85)

print(f"s1 == s3: {s1 == s3}")
print(f"s1 < s2: {s1 < s2}")
print(f"s2 > s1: {s2 > s1}")
print(f"s1 <= s3: {s1 <= s3}")  # __eq__ì™€ __lt__ë¡œ ìë™ ìƒì„±!
print(f"s2 >= s1: {s2 >= s1}")  # ìë™ ìƒì„±!
````

### 2.4 ì‚°ìˆ  ì—°ì‚°ì Special Methods

#### Arithmetic Operators

| ì—°ì‚°ì | Special Method | ì„¤ëª… |
|--------|---------------|------|
| `+` | `__add__(self, other)` | ë§ì…ˆ |
| `-` | `__sub__(self, other)` | ëº„ì…ˆ |
| `*` | `__mul__(self, other)` | ê³±ì…ˆ |
| `/` | `__truediv__(self, other)` | ë‚˜ëˆ—ì…ˆ |
| `//` | `__floordiv__(self, other)` | ëª« |
| `%` | `__mod__(self, other)` | ë‚˜ë¨¸ì§€ |
| `**` | `__pow__(self, other)` | ê±°ë“­ì œê³± |

````python
# ì‚°ìˆ  ì—°ì‚°ì êµ¬í˜„ (Vector ì˜ˆì œ)
print("=== ì‚°ìˆ  ì—°ì‚°ì Special Methods ===")

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"
    
    def __add__(self, other):
        """ë²¡í„° ë§ì…ˆ"""
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        """ë²¡í„° ëº„ì…ˆ"""
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
        """ìŠ¤ì¹¼ë¼ ê³±ì…ˆ"""
        return Vector(self.x * scalar, self.y * scalar)
    
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

v1 = Vector(1, 2)
v2 = Vector(3, 4)

print(f"v1 + v2 = {v1 + v2}")
print(f"v2 - v1 = {v2 - v1}")
print(f"v1 * 3 = {v1 * 3}")
print(f"v1 == v2: {v1 == v2}")
````

### 2.5 Container Special Methods

#### Container Protocol

| Method | ìš©ë„ | í˜¸ì¶œ ë°©ë²• |
|--------|------|----------|
| `__len__(self)` | ê¸¸ì´ ë°˜í™˜ | `len(obj)` |
| `__getitem__(self, key)` | í•­ëª© ê°€ì ¸ì˜¤ê¸° | `obj[key]` |
| `__setitem__(self, key, value)` | í•­ëª© ì„¤ì • | `obj[key] = value` |
| `__delitem__(self, key)` | í•­ëª© ì‚­ì œ | `del obj[key]` |
| `__contains__(self, item)` | í¬í•¨ ì—¬ë¶€ | `item in obj` |
| `__iter__(self)` | ë°˜ë³µì ë°˜í™˜ | `for x in obj` |

````python
# Container Special Methods
print("=== Container Special Methods ===")

class MyList:
    def __init__(self):
        self.items = []
    
    def __len__(self):
        """len() í•¨ìˆ˜ ì§€ì›"""
        return len(self.items)
    
    def __getitem__(self, index):
        """ì¸ë±ì‹± ì§€ì›"""
        return self.items[index]
    
    def __setitem__(self, index, value):
        """í•­ëª© í• ë‹¹ ì§€ì›"""
        self.items[index] = value
    
    def __contains__(self, item):
        """in ì—°ì‚°ì ì§€ì›"""
        return item in self.items
    
    def append(self, item):
        self.items.append(item)
    
    def __repr__(self):
        return f"MyList({self.items})"

my_list = MyList()
my_list.append(1)
my_list.append(2)
my_list.append(3)

print(f"my_list: {my_list}")
print(f"len(my_list): {len(my_list)}")
print(f"my_list[0]: {my_list[0]}")
print(f"2 in my_list: {2 in my_list}")
print(f"5 in my_list: {5 in my_list}")

my_list[0] = 10
print(f"\nAfter my_list[0] = 10: {my_list}")
````

### 2.6 ê¸°íƒ€ ìœ ìš©í•œ Special Methods

#### `__call__`: ê°ì²´ë¥¼ í•¨ìˆ˜ì²˜ëŸ¼ í˜¸ì¶œ
- ê°ì²´ë¥¼ í•¨ìˆ˜ì²˜ëŸ¼ ì‚¬ìš©í•  ìˆ˜ ìˆê²Œ í•¨
- `obj(args)`ì²˜ëŸ¼ í˜¸ì¶œ ê°€ëŠ¥

#### `__enter__`, `__exit__`: Context Manager
- `with` ë¬¸ì—ì„œ ì‚¬ìš©
- ë¦¬ì†ŒìŠ¤ ê´€ë¦¬ (íŒŒì¼, ë„¤íŠ¸ì›Œí¬ ì—°ê²° ë“±)

### 3.5 Abstract Base Class (ì¶”ìƒ ê¸°ë³¸ í´ë˜ìŠ¤)

#### ABCë€?
- **ì •ì˜**: ì§ì ‘ ì¸ìŠ¤í„´ìŠ¤í™”í•  ìˆ˜ ì—†ê³ , ìƒì†ìš©ìœ¼ë¡œë§Œ ì‚¬ìš©ë˜ëŠ” í´ë˜ìŠ¤
- **ìš©ë„**: ì¸í„°í˜ì´ìŠ¤ë¥¼ ì •ì˜í•˜ê³  ê°•ì œ
- **ëª¨ë“ˆ**: `abc` (Abstract Base Classes)

**í•µì‹¬ ìš”ì†Œ**:
- `ABC` í´ë˜ìŠ¤ë¥¼ ìƒì†
- `@abstractmethod` ë°ì½”ë ˆì´í„°ë¡œ ì¶”ìƒ ë©”ì„œë“œ ì •ì˜
- ìì‹ í´ë˜ìŠ¤ì—ì„œ ë°˜ë“œì‹œ êµ¬í˜„í•´ì•¼ í•¨

````python
# Abstract Base Class
from abc import ABC, abstractmethod

print("=== Abstract Base Class ===")

class Shape(ABC):
    """ì¶”ìƒ ê¸°ë³¸ í´ë˜ìŠ¤"""
    
    @abstractmethod
    def area(self):
        """ë„“ì´ë¥¼ ê³„ì‚°í•˜ëŠ” ì¶”ìƒ ë©”ì„œë“œ"""
        pass
    
    @abstractmethod
    def perimeter(self):
        """ë‘˜ë ˆë¥¼ ê³„ì‚°í•˜ëŠ” ì¶”ìƒ ë©”ì„œë“œ"""
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14159 * self.radius ** 2
    
    def perimeter(self):
        return 2 * 3.14159 * self.radius

# ShapeëŠ” ì§ì ‘ ì¸ìŠ¤í„´ìŠ¤í™” ë¶ˆê°€
# shape = Shape()  # TypeError ë°œìƒ!

# ìì‹ í´ë˜ìŠ¤ëŠ” ì¸ìŠ¤í„´ìŠ¤í™” ê°€ëŠ¥
rect = Rectangle(5, 3)
circle = Circle(4)

print(f"ì§ì‚¬ê°í˜• ë„“ì´: {rect.area()}")
print(f"ì§ì‚¬ê°í˜• ë‘˜ë ˆ: {rect.perimeter()}")
print(f"ì› ë„“ì´: {circle.area():.2f}")
print(f"ì› ë‘˜ë ˆ: {circle.perimeter():.2f}")
````

### 4.1 ìº¡ìŠí™”ë€?

- **ì •ì˜**: ë°ì´í„°(ì†ì„±)ì™€ ë©”ì„œë“œë¥¼ í•˜ë‚˜ë¡œ ë¬¶ê³ , ì™¸ë¶€ ì ‘ê·¼ì„ ì œí•œí•˜ëŠ” ê²ƒ

  1. ë°ì´í„° ë³´í˜¸ (ë¬´ë¶„ë³„í•œ ì ‘ê·¼/ìˆ˜ì • ë°©ì§€)

**ì ‘ê·¼ ì œì–´ì** (Pythonì˜ ê´€ë¡€):

**ì¤‘ìš”**: Pythonì—ëŠ” ì§„ì •í•œ privateì´ ì—†ìŒ! ëª¨ë‘ ê´€ë¡€(convention)ì¼ ë¿

````python
# ì ‘ê·¼ ì œì–´ì
print("=== ì ‘ê·¼ ì œì–´ì ===")

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner           # public
        self._balance = balance      # protected (ê´€ë¡€)
        self.__pin = "1234"          # private (Name Mangling)
    
    def deposit(self, amount):
        """ì…ê¸ˆ"""
        if amount > 0:
            self._balance += amount
            print(f"{amount}ì› ì…ê¸ˆ ì™„ë£Œ")
    
    def withdraw(self, amount, pin):
        """ì¶œê¸ˆ"""
        if pin != self.__pin:
            print("PINì´ ì˜¬ë°”ë¥´ì§€ ì•ŠìŠµë‹ˆë‹¤")
            return
        
        if amount > self._balance:
            print("ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤")
            return
        
        self._balance -= amount
        print(f"{amount}ì› ì¶œê¸ˆ ì™„ë£Œ")
    
    def get_balance(self):
        """ì”ì•¡ ì¡°íšŒ"""
        return self._balance

account = BankAccount("ê¹€ì² ìˆ˜", 10000)

# public ì ‘ê·¼ - ê°€ëŠ¥
print(f"ì˜ˆê¸ˆì£¼: {account.owner}")

# protected ì ‘ê·¼ - ê°€ëŠ¥í•˜ì§€ë§Œ ê¶Œì¥í•˜ì§€ ì•ŠìŒ
print(f"ì”ì•¡ (ì§ì ‘ ì ‘ê·¼): {account._balance}")

# private ì ‘ê·¼ - Name Manglingìœ¼ë¡œ ì¸í•´ ë¶ˆê°€
# print(account.__pin)  # AttributeError

# ë©”ì„œë“œë¥¼ í†µí•œ ì ‘ê·¼ - ê¶Œì¥
print(f"ì”ì•¡ (ë©”ì„œë“œ): {account.get_balance()}ì›")
account.deposit(5000)
account.withdraw(3000, "1234")
````

````python
# Property ê¸°ë³¸ ì‚¬ìš©
print("=== Property ê¸°ë³¸ ===")

class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius
    
    @property
    def celsius(self):
        """ì„­ì”¨ ì˜¨ë„ Getter"""
        print("celsius getter í˜¸ì¶œ")
        return self._celsius
    
    @celsius.setter
    def celsius(self, value):
        """ì„­ì”¨ ì˜¨ë„ Setter (ê°’ ê²€ì¦ í¬í•¨)"""
        print(f"celsius setter í˜¸ì¶œ: {value}")
        if value < -273.15:
            raise ValueError("ì ˆëŒ€ ì˜ë„ë³´ë‹¤ ë‚®ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
        self._celsius = value
    
    @property
    def fahrenheit(self):
        """í™”ì”¨ ì˜¨ë„ (ê³„ì‚°ëœ ì†ì„±)"""
        return self._celsius * 9/5 + 32
    
    @fahrenheit.setter
    def fahrenheit(self, value):
        """í™”ì”¨ ì˜¨ë„ ì„¤ì •"""
        self._celsius = (value - 32) * 5/9

# Propertyë¥¼ ì†ì„±ì²˜ëŸ¼ ì‚¬ìš©
temp = Temperature(25)
print(f"í˜„ì¬ ì˜¨ë„: {temp.celsius}Â°C")  # Getter í˜¸ì¶œ
print(f"í™”ì”¨ ì˜¨ë„: {temp.fahrenheit:.1f}Â°F")

# Setter ì‚¬ìš©
temp.celsius = 30  # Setter í˜¸ì¶œ
print(f"ë³€ê²½ëœ ì˜¨ë„: {temp.celsius}Â°C")

# ê°’ ê²€ì¦
try:
    temp.celsius = -300  # ì ˆëŒ€ ì˜ë„ë³´ë‹¤ ë‚®ìŒ
except ValueError as e:
    print(f"\nì—ëŸ¬: {e}")
````

````python
# Getter/Setter vs Property ë¹„êµ
print("=== Getter/Setter vs Property ===")

# 1. Getter/Setter ë°©ì‹
class OldStyleCircle:
    def __init__(self, radius):
        self._radius = radius
    
    def get_radius(self):
        return self._radius
    
    def set_radius(self, radius):
        if radius < 0:
            raise ValueError("ë°˜ì§€ë¦„ì€ 0 ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤")
        self._radius = radius
    
    def get_area(self):
        return 3.14159 * self._radius ** 2

# 2. Property ë°©ì‹
class NewStyleCircle:
    def __init__(self, radius):
        self._radius = radius
    
    @property
    def radius(self):
        return self._radius
    
    @radius.setter
    def radius(self, radius):
        if radius < 0:
            raise ValueError("ë°˜ì§€ë¦„ì€ 0 ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤")
        self._radius = radius
    
    @property
    def area(self):
        return 3.14159 * self._radius ** 2

print("Old Style (Getter/Setter):")
c1 = OldStyleCircle(5)
print(f"  ë°˜ì§€ë¦„: {c1.get_radius()}")
print(f"  ë„“ì´: {c1.get_area():.2f}")
c1.set_radius(10)
print(f"  ë³€ê²½ í›„ ë°˜ì§€ë¦„: {c1.get_radius()}")

print("\nNew Style (Property):")
c2 = NewStyleCircle(5)
print(f"  ë°˜ì§€ë¦„: {c2.radius}")
print(f"  ë„“ì´: {c2.area:.2f}")
c2.radius = 10
print(f"  ë³€ê²½ í›„ ë°˜ì§€ë¦„: {c2.radius}")

print("\nâ†’ Propertyê°€ ë” ê°„ê²°í•˜ê³  Pythonic!")
````

````python
# ì¢…í•© ì˜ˆì œ: ì€í–‰ ê³„ì¢Œ ì‹œìŠ¤í…œ
from abc import ABC, abstractmethod
from datetime import datetime

print("=== ì€í–‰ ê³„ì¢Œ ì‹œìŠ¤í…œ (ì¢…í•© ì˜ˆì œ) ===")

class Account(ABC):
    """ê³„ì¢Œ ì¶”ìƒ ê¸°ë³¸ í´ë˜ìŠ¤"""
    
    account_count = 0  # Class Attribute
    
    def __init__(self, owner, balance=0):
        self.owner = owner
        self._balance = balance
        self.__pin = "1234"  # Private
        Account.account_count += 1
        self.account_number = f"ACC{Account.account_count:04d}"
        self.transactions = []
    
    # Property
    @property
    def balance(self):
        return self._balance
    
    # Abstract Method
    @abstractmethod
    def calculate_interest(self):
        pass
    
    # Instance Method
    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("ì…ê¸ˆì•¡ì€ 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤")
        self._balance += amount
        self._add_transaction("ì…ê¸ˆ", amount)
        return self._balance
    
    def withdraw(self, amount, pin):
        if pin != self.__pin:
            raise ValueError("PINì´ ì˜¬ë°”ë¥´ì§€ ì•ŠìŠµë‹ˆë‹¤")
        if amount > self._balance:
            raise ValueError("ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤")
        self._balance -= amount
        self._add_transaction("ì¶œê¸ˆ", -amount)
        return self._balance
    
    def _add_transaction(self, type_, amount):
        """Protected Method: ê±°ë˜ ë‚´ì—­ ì¶”ê°€"""
        self.transactions.append({
            'type': type_,
            'amount': amount,
            'balance': self._balance,
            'time': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        })
    
    # Class Method
    @classmethod
    def get_account_count(cls):
        return cls.account_count
    
    # Special Methods
    def __str__(self):
        return f"{self.owner}ì˜ {self.account_number} (ì”ì•¡: {self._balance:,}ì›)"
    
    def __repr__(self):
        return f"{self.__class__.__name__}('{self.owner}', {self._balance})"
    
    def __eq__(self, other):
        return self.account_number == other.account_number
    
    def __lt__(self, other):
        return self._balance < other._balance

class SavingsAccount(Account):
    """ì˜ˆê¸ˆ ê³„ì¢Œ"""
    
    def __init__(self, owner, balance=0, interest_rate=0.02):
        super().__init__(owner, balance)
        self.interest_rate = interest_rate
    
    def calculate_interest(self):
        """ì´ì ê³„ì‚° ë° ì ìš©"""
        interest = self._balance * self.interest_rate
        self._balance += interest
        self._add_transaction("ì´ì", interest)
        return interest

class CheckingAccount(Account):
    """ì…ì¶œê¸ˆ ê³„ì¢Œ (ë§ˆì´ë„ˆìŠ¤ í†µì¥)"""
    
    def __init__(self, owner, balance=0, overdraft_limit=1000000):
        super().__init__(owner, balance)
        self.overdraft_limit = overdraft_limit
    
    def calculate_interest(self):
        """ë§ˆì´ë„ˆìŠ¤ ì”ì•¡ì— ëŒ€í•œ ì´ì ë¶€ê³¼"""
        if self._balance < 0:
            interest = abs(self._balance) * 0.05  # 5% ì´ì ë¶€ê³¼
            self._balance -= interest
            self._add_transaction("ì´ì ë¶€ê³¼", -interest)
            return -interest
        return 0
    
    def withdraw(self, amount, pin):
        """ì˜¤ë²„ë¼ì´ë”©: ë§ˆì´ë„ˆìŠ¤ í•œë„ í™•ì¸"""
        if amount > self._balance + self.overdraft_limit:
            raise ValueError("ë§ˆì´ë„ˆìŠ¤ í•œë„ë¥¼ ì´ˆê³¼í–ˆìŠµë‹ˆë‹¤")
        return super().withdraw(amount, pin)

# ì‚¬ìš© ì˜ˆì‹œ
print("\n1. ê³„ì¢Œ ìƒì„±")
savings = SavingsAccount("ê¹€ì² ìˆ˜", 1000000, 0.03)
checking = CheckingAccount("ì´ì˜í¬", 500000, 2000000)

print(f"ì˜ˆê¸ˆ ê³„ì¢Œ: {savings}")
print(f"ì…ì¶œê¸ˆ ê³„ì¢Œ: {checking}")
print(f"ì´ ê³„ì¢Œ ìˆ˜: {Account.get_account_count()}")

print("\n2. ì…ì¶œê¸ˆ")
savings.deposit(500000)
print(f"ì…ê¸ˆ í›„: {savings}")

checking.withdraw(700000, "1234")
print(f"ì¶œê¸ˆ í›„: {checking}")

print("\n3. ì´ì ê³„ì‚°")
interest = savings.calculate_interest()
print(f"ì˜ˆê¸ˆ ì´ì: {interest:,.0f}ì›")
print(f"ì´ì í›„: {savings}")

print("\n4. ê³„ì¢Œ ë¹„êµ")
print(f"savings > checking: {savings > checking}")

print("\n5. ê±°ë˜ ë‚´ì—­ (ìµœê·¼ 3ê±´)")
for trans in savings.transactions[-3:]:
    print(f"  {trans['time']}: {trans['type']} {trans['amount']:,.0f}ì› (ì”ì•¡: {trans['balance']:,.0f}ì›)")
````
