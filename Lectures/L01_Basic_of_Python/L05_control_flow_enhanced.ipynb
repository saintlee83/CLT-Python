### 2.3 match/case (Python 3.10+)

**êµ¬ì¡°ì  íŒ¨í„´ ë§¤ì¹­**:
- ì—¬ëŸ¬ ê²½ìš°ë¥¼ ê¹”ë”í•˜ê²Œ ì²˜ë¦¬
- íŒ¨í„´ ë§¤ì¹­ ì§€ì›
- ê°€ë…ì„± í–¥ìƒ

**íŒ¨í„´ ì¢…ë¥˜**:
1. ë¦¬í„°ëŸ´ íŒ¨í„´: êµ¬ì²´ì ì¸ ê°’
2. ìº¡ì²˜ íŒ¨í„´: ë³€ìˆ˜ì— ê°’ ì €ì¥

````python
# match/case
print("=== match/case ===")
print()

# ê¸°ë³¸ ì‚¬ìš©
command = "start"

match command:
    case "start":
        print("ì‹œì‘í•©ë‹ˆë‹¤")
    case "stop":
        print("ì¤‘ì§€í•©ë‹ˆë‹¤")
    case "pause":
        print("ì¼ì‹œì •ì§€í•©ë‹ˆë‹¤")
    case _:
        print("ì•Œ ìˆ˜ ì—†ëŠ” ëª…ë ¹")
print()

# íŒ¨í„´ ë§¤ì¹­
point = (0, 0)

match point:
    case (0, 0):
        print("ì›ì ")
    case (0, y):
        print(f"Yì¶• ìœ„ì˜ ì : {y}")
    case (x, 0):
        print(f"Xì¶• ìœ„ì˜ ì : {x}")
    case (x, y):
        print(f"ì¼ë°˜ ì : ({x}, {y})")
````

### 2.4 Guard Clause íŒ¨í„´

**Guard Clause**: ì¡°ê±´ì„ ë¨¼ì € ê²€ì‚¬í•˜ì—¬ ì¡°ê¸° ë°˜í™˜

**ì¥ì **:
- ì¤‘ì²© ê°ì†Œ (Cyclomatic Complexity ë‚®ì¶¤)
- ê°€ë…ì„± í–¥ìƒ (í•´í”¼ íŒ¨ìŠ¤ê°€ ë§ˆì§€ë§‰)
- ì˜ˆì™¸ ìƒí™© ë¨¼ì € ì²˜ë¦¬

**ì›ì¹™**:
1. ì˜ˆì™¸/ì—ëŸ¬ ì¼€ì´ìŠ¤ë¥¼ ë¨¼ì € ê²€ì‚¬
2. ì¡°ê±´ ë§Œì¡±í•˜ì§€ ì•Šìœ¼ë©´ ì¦‰ì‹œ ë°˜í™˜

````python
# Guard Clause
print("=== Guard Clause ===")
print()

# âŒ ë‚˜ìœ ì˜ˆ (ì¤‘ì²© ê¹ŠìŒ)
def process_data_bad(data):
    if data is not None:
        if len(data) > 0:
            if isinstance(data, list):
                return sum(data)
            else:
                return 0
        else:
            return 0
    else:
        return 0

# âœ… ì¢‹ì€ ì˜ˆ (Guard Clause)
def process_data_good(data):
    # ì˜ˆì™¸ ìƒí™© ë¨¼ì € ì²˜ë¦¬
    if data is None:
        return 0
    if len(data) == 0:
        return 0
    if not isinstance(data, list):
        return 0
    
    # ì •ìƒ ë¡œì§
    return sum(data)

# í…ŒìŠ¤íŠ¸
print(f"None: {process_data_good(None)}")
print(f"ë¹ˆ ë¦¬ìŠ¤íŠ¸: {process_data_good([])}")
print(f"ì •ìƒ: {process_data_good([1, 2, 3])}")
````
