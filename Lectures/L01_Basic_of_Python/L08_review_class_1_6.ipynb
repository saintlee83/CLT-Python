## ğŸ• Class 1-2 ì¬í•´ì„: Tokenê³¼ Namespace

### Token ë‹¨ê³„ì—ì„œ ì´ë¯¸ Namespaceê°€ ê²°ì •ëœë‹¤

**Identifier (ì‹ë³„ì) Token** â†’ Namespaceì—ì„œ ì°¾ì„ ì´ë¦„

````python
# Class 1-2 ì¬í•´ì„
import tokenize
import io

print("=== Tokenê³¼ Namespace ===")

code = "x = 10 + y"
tokens = tokenize.generate_tokens(io.StringIO(code).readline)

print(f"ì½”ë“œ: {code}")
print("\nToken ë¶„ì„:")
for token in tokens:
    if token.type == tokenize.NAME:
        print(f"  NAME '{token.string}' â†’ Namespaceì—ì„œ ì°¾ì„ ì´ë¦„")
    elif token.type == tokenize.NUMBER:
        print(f"  NUMBER '{token.string}' â†’ ì§ì ‘ì ì¸ ê°ì²´ ìƒì„±")
    elif token.type == tokenize.OP:
        print(f"  OP '{token.string}' â†’ ì—°ì‚° ìˆ˜í–‰")
````

## ğŸ•‘ Class 3 ì¬í•´ì„: Expressionê³¼ Name Resolution

### Expression í‰ê°€ = Namespaceì—ì„œ ì´ë¦„ ì°¾ê¸°

**Expressionì´ í‰ê°€ë  ë•Œ LEGB ë£°ì´ ì‘ë™**

````python
# Class 3 ì¬í•´ì„: Expressionê³¼ Namespace
print("=== Expression í‰ê°€ì™€ LEGB ===")

x = 100  # Global

def outer():
    x = 50  # Enclosing
    
    def inner():
        # xë¥¼ Expressionì—ì„œ ì‚¬ìš©
        result = x + 10  # ì–´ë–¤ x?
        print(f"inner - result = {result}")
        print("  â†’ Localì— x ì—†ìŒ")
        print("  â†’ Enclosingì—ì„œ x ë°œê²¬: 50")
        print("  â†’ 50 + 10 = 60")
    
    inner()

outer()

print("\nLEGB ìˆœì„œë¡œ xë¥¼ ì°¾ìŒ:")
print("  L (Local) â†’ ì—†ìŒ")
print("  E (Enclosing) â†’ 50 ë°œê²¬! âœ…")
````

## ğŸ•” Class 6 ì¬í•´ì„: Iteratorì™€ Namespace

### Generatorì™€ Closure = Namespace í™œìš©ì˜ ê·¹ì¹˜

**GeneratorëŠ” ìì‹ ë§Œì˜ Namespaceë¥¼ ìœ ì§€**

````python
# Class 6 ì¬í•´ì„: Generatorì™€ Namespace
print("=== Generatorì™€ Namespace ===")

def counter():
    count = 0  # Generatorì˜ Local Namespace
    while True:
        count += 1
        yield count

# Generator ê°ì²´ ìƒì„±
c1 = counter()
c2 = counter()

print("c1:", next(c1), next(c1), next(c1))
print("c2:", next(c2), next(c2))

print("\ní•´ì„:")
print("  ê° GeneratorëŠ” ë…ë¦½ì ì¸ Namespace ìœ ì§€")
print("  countëŠ” ê° Generator ë‚´ì—ì„œë§Œ ì¡´ì¬")
print("  yield ì‚¬ì´ì— ìƒíƒœ(Namespace) ë³´ì¡´")
````

````python
# ì¢…í•© ì˜ˆì œ: ëª¨ë“  ê°œë… í†µí•©
print("=== ì¢…í•© ì˜ˆì œ ===")

# Global Namespace
counter = 0

def outer(x):
    # Enclosing Namespace
    multiplier = 2
    
    def inner(y):
        # Local Namespace
        global counter
        counter += 1
        
        # Name Resolution: LEGB
        result = x * multiplier + y  # x(param), multiplier(enclosing), y(param)
        
        # Built-in
        return len(str(result))  # lenì€ Built-in
    
    return inner

# ì‹¤í–‰
func = outer(10)
print(f"ê²°ê³¼: {func(5)}")
print(f"counter: {counter}")

print("\në¶„ì„:")
print("  x=10      â†’ outerì˜ Parameter (Enclosing)")
print("  multiplier â†’ Enclosing Namespace")
print("  y=5       â†’ innerì˜ Parameter (Local)")
print("  counter   â†’ Global Namespace (global ì„ ì–¸)")
print("  len       â†’ Built-in Namespace")
print("  str       â†’ Built-in Namespace")
````

````python
# Class 7 ë¯¸ë¦¬ë³´ê¸°
print("=== Class 7 ë¯¸ë¦¬ë³´ê¸° ===")

# 1. Closure = Enclosing Namespace í¬íš
def make_multiplier(n):
    def multiplier(x):
        return x * n  # nì€ Enclosingì—ì„œ í¬íš
    return multiplier

times2 = make_multiplier(2)
times3 = make_multiplier(3)

print(f"times2(5) = {times2(5)}")
print(f"times3(5) = {times3(5)}")
print("â†’ ê° í•¨ìˆ˜ëŠ” ìì‹ ë§Œì˜ 'n'ì„ ê¸°ì–µ")

# 2. nonlocalë¡œ Enclosing ìˆ˜ì •
def counter():
    count = 0
    
    def increment():
        nonlocal count  # Enclosingì˜ count ìˆ˜ì •
        count += 1
        return count
    
    return increment

c = counter()
print(f"\n{c()}, {c()}, {c()}")
print("â†’ nonlocalë¡œ Enclosing Namespace ìˆ˜ì •")
````
