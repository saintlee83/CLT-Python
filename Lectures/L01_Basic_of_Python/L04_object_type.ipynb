#### 3) Identity (ì •ì²´ì„±)

**ğŸ“– Identityë€?**
- **ì •ì˜**: ê°ì²´ì˜ ë©”ëª¨ë¦¬ ì£¼ì†Œ (ê³ ìœ  ì‹ë³„ì)
- **í™•ì¸**: `id()` í•¨ìˆ˜
- **ë¹„êµ**: `is` ì—°ì‚°ì

**íŠ¹ì§•**:
- ê°ì²´ë§ˆë‹¤ ê³ ìœ í•œ ê°’
- ê°ì²´ ìƒì„± ì‹œ ê²°ì •, ë³€ê²½ ë¶ˆê°€
- `is` ì—°ì‚°ìë¡œ ë¹„êµ (ê°™ì€ ê°ì²´ì¸ì§€)

**is vs ==**:
```python
is  : ê°™ì€ ê°ì²´ì¸ê°€? (identity ë¹„êµ)
==  : ê°™ì€ ê°’ì¸ê°€?   (value ë¹„êµ)
```

````python
# Identity í™•ì¸
print("=== Objectì˜ Identity ===")
print()

# ê°™ì€ ê°’, ë‹¤ë¥¸ ê°ì²´
a = [1, 2, 3]
b = [1, 2, 3]
print(f"a = [1, 2, 3], id: {id(a)}")
print(f"b = [1, 2, 3], id: {id(b)}")
print(f"a == b: {a == b}  (ê°’ ë¹„êµ)")
print(f"a is b: {a is b}  (ì •ì²´ì„± ë¹„êµ)")
print()

# ê°™ì€ ê°ì²´ ê°€ë¦¬í‚¤ê¸°
c = a
print(f"c = a")
print(f"cì˜ id: {id(c)}")
print(f"a is c: {a is c}  (ê°™ì€ ê°ì²´!)")
print()

# íŠ¹ìˆ˜ ì¼€ì´ìŠ¤: ì‘ì€ ì •ìˆ˜ëŠ” ìºì‹±
x = 10
y = 10
print(f"x = 10, y = 10")
print(f"x is y: {x is y}  (ì‘ì€ ì •ìˆ˜ëŠ” ìºì‹±ë¨)")

x = 1000
y = 1000
print(f"x = 1000, y = 1000")
print(f"x is y: {x is y}  (í° ì •ìˆ˜ëŠ” ë‹¤ë¥¸ ê°ì²´)")
````

#### 4) Reference Count (ì°¸ì¡° íšŸìˆ˜)

**ğŸ“– Reference Countë€?**
- **ì •ì˜**: ê°ì²´ë¥¼ ê°€ë¦¬í‚¤ëŠ” ë³€ìˆ˜/ì°¸ì¡°ì˜ ê°œìˆ˜
- **í™•ì¸**: `sys.getrefcount()` í•¨ìˆ˜
- **ì—­í• **: ë©”ëª¨ë¦¬ ê´€ë¦¬ (Garbage Collection)

**íŠ¹ì§•**:
- ì°¸ì¡° íšŸìˆ˜ê°€ 0ì´ ë˜ë©´ ë©”ëª¨ë¦¬ í•´ì œ
- Pythonì˜ ìë™ ë©”ëª¨ë¦¬ ê´€ë¦¬ ê¸°ë³¸ ë©”ì»¤ë‹ˆì¦˜
- ê°œë°œìê°€ ì§ì ‘ ê´€ë¦¬í•  í•„ìš” ì—†ìŒ

**Reference Count ë³€í™”**:
```python
x = [1, 2, 3]  # refcount = 1
y = x          # refcount = 2 (ì¦ê°€)
del y          # refcount = 1 (ê°ì†Œ)
del x          # refcount = 0 â†’ ë©”ëª¨ë¦¬ í•´ì œ
```

````python
# Reference Count í™•ì¸
import sys

print("=== Reference Count ===")
print()

x = [1, 2, 3]
print(f"x = [1, 2, 3]")
print(f"refcount: {sys.getrefcount(x) - 1}")  # -1: getrefcount ìì²´ì˜ ì°¸ì¡° ì œì™¸
print()

y = x  # ì°¸ì¡° ì¦ê°€
print(f"y = x")
print(f"refcount: {sys.getrefcount(x) - 1}")
print()

z = x  # ì°¸ì¡° ì¦ê°€
print(f"z = x")
print(f"refcount: {sys.getrefcount(x) - 1}")
print()

del y  # ì°¸ì¡° ê°ì†Œ
print(f"del y")
print(f"refcount: {sys.getrefcount(x) - 1}")
print()

del z  # ì°¸ì¡° ê°ì†Œ
print(f"del z")
print(f"refcount: {sys.getrefcount(x) - 1}")
````

### 1.3 First-class Object

#### ğŸ“– First-class Objectë€?

**ì •ì˜**: ë‹¤ë¥¸ ê°ì²´ë“¤ì´ ì¼ë°˜ì ìœ¼ë¡œ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” ëª¨ë“  ì—°ì‚°ì„ ì§€ì›í•˜ëŠ” ê°ì²´

**First-classì˜ ì¡°ê±´**:
1. ë³€ìˆ˜ì— í• ë‹¹ ê°€ëŠ¥
2. í•¨ìˆ˜ì˜ ì¸ìë¡œ ì „ë‹¬ ê°€ëŠ¥
3. í•¨ìˆ˜ì˜ ë°˜í™˜ê°’ìœ¼ë¡œ ì‚¬ìš© ê°€ëŠ¥
4. ìë£Œêµ¬ì¡°ì— ì €ì¥ ê°€ëŠ¥

**Pythonì˜ First-class Object**:
- í•¨ìˆ˜ (Function)
- í´ë˜ìŠ¤ (Class)
- ëª¨ë“ˆ (Module)

**ì™œ ì¤‘ìš”í•œê°€?**
- ìœ ì—°í•œ í”„ë¡œê·¸ë˜ë° ê°€ëŠ¥
- ê³ ì°¨ í•¨ìˆ˜ (Higher-order Function) êµ¬í˜„
- Decorator, Closure ë“± ê³ ê¸‰ íŒ¨í„´ ì‚¬ìš©

````python
# First-class Function
print("=== First-class Function ===")
print()

def square(x):
    return x ** 2

def cube(x):
    return x ** 3

# 1. ë³€ìˆ˜ì— í• ë‹¹
f = square
print(f"1. ë³€ìˆ˜ì— í• ë‹¹: f = square")
print(f"   f(5) = {f(5)}")
print()

# 2. í•¨ìˆ˜ì˜ ì¸ìë¡œ ì „ë‹¬
def apply(func, value):
    return func(value)

print(f"2. ì¸ìë¡œ ì „ë‹¬: apply(square, 5)")
print(f"   ê²°ê³¼: {apply(square, 5)}")
print()

# 3. í•¨ìˆ˜ì˜ ë°˜í™˜ê°’
def get_operation(op):
    if op == 'square':
        return square
    else:
        return cube

operation = get_operation('square')
print(f"3. ë°˜í™˜ê°’: operation = get_operation('square')")
print(f"   operation(5) = {operation(5)}")
print()

# 4. ìë£Œêµ¬ì¡°ì— ì €ì¥
operations = [square, cube]
print(f"4. ë¦¬ìŠ¤íŠ¸ì— ì €ì¥: operations = [square, cube]")
print(f"   operations[0](5) = {operations[0](5)}")
print(f"   operations[1](5) = {operations[1](5)}")
````
