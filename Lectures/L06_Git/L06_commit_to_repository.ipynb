# git commit

stage 에서 repository로 이동 시켜 version 관리를 실제로 시작하게 하는 명령어

snapshot 방식
- 현재 repository에 저장된 HEAD와 stage의 내용의 차이를 비교하여 차이점만을 저장하는 방식
- 차이점을 사진을 찍듯이 저장한다는 의미

## git commit -m "message"

```
git commit -m "커밋 메세지"
```

-m : 로그를 inline으로 편집하는 옵션

### lifecycle

1. working tree에서 파일 수정
2. git add -> stage로 이동
3. commit -> log와 함께 repository에 저장

## git commit -am "message"

add + commit 을 동시에 진행
- 이미 Tracked 상태의 모든 파일들에 대한 변경사항을 stage에 올리고 commit함
- 새로운 파일(Untracked file)은 해당 안됨
