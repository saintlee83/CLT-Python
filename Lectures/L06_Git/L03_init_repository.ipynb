# git init

기존 로컬 디렉토리를 git repository로 만듦

주의 `!`는 노트북에서 쉘 스크립트를 사용하기 위해서 붙음

```python
!mkdir test_repo
!cd test_repo
!git init -b main
```

`git init -b main` : 기본 브랜치를 main으로 설정

- 기본 브랜치 명은 master였으나 BLM이후 인종 차별 어쩌고 저쩌고...

위의 예에서 first-git은 working tree(or working directory)라고 불리며, 현재 상태의 sanpshot을 남길 수 있음

`git clone` : remote repository(원격 저장소)에서 파일을 가져옴

# git 에서 파일의 논리적 위치

## Working Tree(working directory)

test_repo가 working directory
- `.git`을 제외한 것을 가르킴
- 파일의 수정, 저장 등의 작업이 이루어짐

## Stage Area(or Index, taging area)

- Version 관리될 파일들이 대기하는 논리적인 장소
  - 버전 관리가 되고 있는 파일이나 새로 변경된 경우
  - 새로 추가된 파일

- .git/index에 저장됨
  - index = git 내부의 binary file로 Staging Area의 정보를 가지고 있음

## Repository

- Stage -commit-> repository
- Version이 할당되어 관리되는 파일들이 있음

repository와 stage만 존재하는 저장소(bare repository)를 생성(주로 서버용)하는 것도 가능함: --bare 옵션 사용.

## git life cycle

- 구분 : tracked file, untracked file
- tracked file : unmodified, modified, staged file
