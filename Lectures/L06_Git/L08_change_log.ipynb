# git log : commit history

git 으로 관리되는 프로젝트의 commit history를 보여줌
- 누가 언제 어떻게 프로젝트를 변경했는지, 브랜치간 관계, commit id등을 확인 가능

```shell
git log
```

현재 branch의 commit history를 역순으로 출력
- author, date, commit message, commit id (SHA1 hash value)

```shell
git log <branch_name>
```

### 주요 Options

- `--branches` : 모든 local 브랜치의 commit history를 출력 (remote branch까지 포함하려면 `--all` 사용)

- `--oneline` : 각 commit을 한 줄로 간략하게 출력

- `-n <int>` : 출력할 commit의 개수를 숫자로 지정

- `--stat` : 각 commit에서 변경된 파일과 추가/삭제된 라인 수를 통계로 출력

- `--graph` : 브랜치의 merge 관계를 ASCII 그래프로 시각화하여 출력

## 범위

```shell
git log <s_branch>..<e_branch>
```
- 두 branch의 차이를 출력

- s_branch에 없고 e_branch에 있는 commit 출력

```shell
git log <s_commit>..<e_commit>
```
- 시작 부터 끝사이의 commit history를 출력

- git log HEAD~4..HEAD : HEAD를 기준으로 4번째 전 commit 이후 부터 최신 commit까지의 history를 출력

**Releative Reference**
- 범위를 나타내는 .. 를 사용할 때,

- 시작하는 위치 이후 부터 (exclusive)

- 끝나는 위치까지임 (inclusive)

- (시작 commit id등은 exclusive임)
