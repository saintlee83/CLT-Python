# git add

- untracked file을 stage에 이동 시키는 명령어 -> Tracked file로 변경
- 기존 tracked file인 경우, modified file을 stage에 올리는데에 사용

다음 명령어는 new_file.txt를 stage 상태로 변경함
```
git add new_file.txt
```

- `Changes to be committed:` 아래 표시되는 파일들이 stage에 있는 파일
- untracked file이 add이후 commit할 수 있는 상태가 됨

한번에 여러 파일을 staged로 만드는 법
```
git add .

git add -A # 삭제된 파일까지 Tracked
```
