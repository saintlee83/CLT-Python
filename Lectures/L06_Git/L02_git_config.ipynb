
## Local Repository를 위한 기본 설정

### git config

`git config`는 Git의 동작 방식을 사용자 환경에 맞게 설정하는 명령어입니다.

#### 설정 범위 (Scope)

| 옵션 | 설정 파일 위치 | 적용 범위 |
|------|---------------|----------|
| `--system` | `/etc/gitconfig` | 시스템 전체 (모든 사용자) |
| `--global` | `~/.gitconfig` | 현재 사용자의 모든 저장소 |
| `--local` | `.git/config` | 현재 저장소만 (기본값) |

우선순위: `local` > `global` > `system`

#### 필수 설정: 사용자 정보

Git 커밋에는 작성자 정보가 필수입니다.

```shell
# 사용자 이름 설정
git config --global user.name "Your Name"

# 이메일 설정
git config --global user.email "your.email@example.com"
```

#### 설정 확인

```shell
# 모든 설정 확인
git config --list

# 특정 설정 확인
git config user.name
git config user.email

# 설정이 어느 파일에서 왔는지 확인
git config --list --show-origin
```

#### 설정 삭제

```shell
git config --global --unset user.name
```


#### 유용한 추가 설정

```shell
# 기본 브랜치 이름 설정 (main으로 설정)
git config --global init.defaultBranch main

# 컬러 출력 활성화
git config --global color.ui auto

# 단축 명령어 (alias) 설정
git config --global alias.st status
git config --global alias.co checkout
git config --global alias.br branch
git config --global alias.cm commit
git config --global alias.lg "log --oneline --graph --all"
```


### autocrlf

개행문자 : new line

```
core.autocrlf
```

깃에서 파일을 git repository에 업로드할 때 개행문자를 처리하는 옵션

windows : \r\n <- git -> mac : \n

```
core.autocrlf = false
- 파일 내용을 그대로 반영, 비추천함

core.autocrlf = true
- 모든 개행 문자를 LF(linux, mac)로 변경

core.autocrlf = input
- 업로드시 CRLF가 있으면 LF로 변경
- local은 그대로 반영
```

[CR and LF](https://velog.io/@jykim29/CRCarriage-Return%EA%B3%BC-LFLine-Feed)


### editor

```shell
git config --global core.editor "vim"
git config --global core.editor "code --wait"
git config --global core.editor "nano"
git config --global core.editor "notepad"
```

혹은 `.gitconfig` 에 다음 내용 추가

```text
editor=vim
```


### vi(vim) 사용법

visual mode : 말 그대로, 'h, j, k, l'로 커서 이동
command mode : `:`을 입력하여 명령모드로 진입
nomal mode : 이건 나도 모르겠음
insert mode : 입력 모드, `i`를 입력하여 진입
