# git status

- 현재 working directory의 파일들의 상태를 확인하는데 사용
- -s or --short 옵션을 이용하여 간략한 상태 확인 가능
- git init 이후 사용 가능
- 새로운 파일이 추가되면 untracked file이 있음을 알림


## .gitignore

untracked file : git이 version 관리 자체를 하지 않는 파일
- .gitignore 파일을 통해 격리된 파일들은 모두 untracked file
- 일종의 txt파일임, git으로 부터 격리할 resourcef의 이름 또는 패턴을 리스트로 가짐

### .gitignore 패턴 작성 방법
#### 1. 특정 파일명 지정
```
file_name.file_ext
```
- git으로 관리되는 프로젝트의 모든 하위 디렉토리에서 해당 파일명(확장자 포함)을 가진 resource를 모두 격리

#### 2. 특정 경로의 파일 지정
```
/path/file_name.file_ext
```
- 특정 경로 내의 해당 파일명과 확장자를 가지는 resource를 격리

#### 3. 특정 확장자(extension) 지정
```
*.file_ext
```
- git으로 관리되는 프로젝트의 모든 하위 디렉토리에서 해당 확장자를 가지는 모든 resource를 격리

#### 4. 이름으로 지정
```
file_or_directory_name
```
- 해당 이름을 가지는 모든 파일과 디렉토리를 격리

#### 5. 특정 디렉토리 지정
```
directory_name/
```
- 뒤에 붙은 `/` (slash)가 디렉토리임을 의미하며, 해당 디렉토리 내의 모든 resources를 격리

### 참고: 추가 패턴 규칙

#### 6. 예외 지정 (!)
```
!file_name.file_ext
```
- `!`로 시작하는 패턴은 격리하지 않는 예외를 의미
- 다른 패턴에 의해 격리 대상이더라도 git이 추적하도록 함

#### 7. 와일드카드 (*)
```
*.file_ext
```
- `*`는 0개 이상의 모든 문자를 의미
- 모든 패턴에서 사용 가능

#### 8. 하위 디렉토리 전체 (**)
```
logs/**/*.txt
```
- `**`는 모든 하위 디렉토리를 의미
- 예: `logs/**/*.txt`는 logs 디렉토리의 모든 하위 디렉토리에 있는 .txt 파일을 격리
