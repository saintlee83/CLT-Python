````python
def calculate_mean_with_list_excluding_zeros(arr):
  """0을 제외하고 리스트의 평균을 파이썬 내장 기능만 사용하여 계산합니다."""
  try:
    # 입력이 반복 가능한지 확인
    if not isinstance(arr, (list, tuple)):
      raise TypeError("입력은 리스트 또는 튜플이어야 합니다.")

    non_zero_elements = []
    for x in arr:
      # 요소가 숫자인지 확인
      if not isinstance(x, (int, float)):
        raise TypeError(f"리스트 요소는 숫자여야 하지만, {type(x).__name__} 타입이 발견되었습니다.")
      if x != 0:
        non_zero_elements.append(x)

    if not non_zero_elements:
      # 리스트가 0만 포함하거나 필터링 후 비어 있는 경우 처리
      return 0
    return sum(non_zero_elements) / len(non_zero_elements)
  except TypeError as e:
    raise ValueError(f"잘못된 입력: {e}") from e
  except Exception as e:
    # 기타 예상치 못한 오류 처리
    raise RuntimeError(f"예상치 못한 오류가 발생했습니다: {e}") from e

# 사용 예시:
# my_list_with_zeros = [1, 0, 2, 0, 3, 4, 5]
# mean_value_excluding_zeros = calculate_mean_with_list_excluding_zeros(my_list_with_zeros)
# print(f"0을 제외한 리스트의 평균은: {mean_value_excluding_zeros}")

# 잘못된 입력 처리 예시:
# try:
#   calculate_mean_with_list_excluding_zeros("not a list")
# except ValueError as e:
#   print(e)

# try:
#   calculate_mean_with_list_excluding_zeros([1, 2, 'a'])
# except ValueError as e:
#   print(e)
````

````python
def sort_strings_alphabetically(string_list: list[str]) -> list[str]:
  """문자열 리스트를 입력받아 알파벳 순서로 정렬하여 반환합니다.

  Args:
    string_list: 정렬할 문자열의 리스트입니다.

  Returns:
    알파벳 순서로 정렬된 문자열 리스트입니다.
  """
  return sorted(string_list)

# 예시 사용법:
# my_strings = ["banana", "apple", "cherry", "date"]
# sorted_strings = sort_strings_alphabetically(my_strings)
# print(sorted_strings)
````

## 상관 관계 분석

### Subtask:

**Reasoning**:
The subtask requires calculating and visualizing the correlation matrix for numerical variables. The `df.corr()` method will calculate the correlation matrix, and `seaborn.heatmap()` will be used to visualize it, with `matplotlib.pyplot` to set figure size and title.



````python
import seaborn as sns

correlation_matrix = df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('상관 관계 행렬')
plt.show()
````
